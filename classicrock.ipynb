## Section 6-Histograms: Use the DataFrame's hist() function to create a histogram for each numerical column.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df'
df = pd.read_csv('classic-rock-song-list.csv')

columns_of_interest = ['Song_Clean', 'ARTIST_CLEAN', 'Release_Year', 'COMBINED', 'First?', 'Year?', 'PlayCount', 'F*G']

# Filter out the numerical columns
numerical_columns = df[columns_of_interest].select_dtypes(include=['int64', 'float64']).columns

for col in numerical_columns:
    df[col].hist()
    plt.title(col)
    plt.show()
```

## Section 9-Map: Use map() to transform some of the data. 


```python
import pandas as pd

# Assuming you have a DataFrame named 'df'
df = pd.read_csv('classic-rock-song-list.csv')

columns_of_interest = ['Song_Clean', 'ARTIST_CLEAN', 'Release_Year', 'COMBINED', 'First?', 'Year?', 'PlayCount', 'F*G']

# Convert 'Release_Year' to numeric, errors='coerce' will replace non-numeric values with NaN
df['Release_Year'] = pd.to_numeric(df['Release_Year'], errors='coerce')

# Use map() to transform the 'Release_Year' column
df['Release_Year'] = df['Release_Year'].map(lambda x: x if x >= 2000 else None)

print(df[columns_of_interest])
```

## Tell a Story - Which was the most played Classic Rock Song from the CSV list?

```python
import pandas as pd

# Assuming you have a DataFrame named 'df'
df = pd.read_csv('classic-rock-song-list.csv')

# Use idxmax() to find the index of the row with the highest 'PlayCount'
most_played_index = df['PlayCount'].idxmax()

# Get the 'Song_Clean' and 'ARTIST_CLEAN' of the most played song
most_played_song = df.loc[most_played_index, 'Song_Clean']
most_played_artist = df.loc[most_played_index, 'ARTIST_CLEAN']

print(f"The most played song is '{most_played_song}' by {most_played_artist}.")
```
